```python
import numpy as np
from xai.pca import PCAAnalysis
```

```python
dataset  = "TkaFewShotCompleteDataset/images"
#dataset  = "KL01_KL234_Real/KL234"
synth_dataset = "Bilateral_Knee_XRay"
#synth_dataset = "KL234"
real_images = f"/data_vault/hexai/{dataset}/**"
fake_images = f"/data_vault/hexai/SyntheticKneeImages/{synth_dataset}/**"
```

```python
pca_analysis = PCAAnalysis(real_images, fake_images, n_components=2)
```

```python
pca_analysis.fit()
```

```python
pca_analysis.statistical_significance_test(first_n_components=2)
```

```python
pca_analysis.visualize_components()
```

```python
pca_analysis.pc_score_analysis(first_n_components=2)
```

```python
# KL234 Mask
outliers = np.where( ((pca_analysis.real_scores[:, 0] > -300) & ((pca_analysis.real_scores[:, 0] < 300)) & ((pca_analysis.real_scores[:, 1] < 200) & (pca_analysis.real_scores[:, 1] > -200))))[0]
```

```python
pca_analysis.pc_score_analysis(first_n_components=2)
```

```python
pca_analysis.refit_without_outliers(outliers)
```

```python
pca_analysis.visualize_components()
```
